```python
import cv2 
```

```python
import matplotlib.pyplot as plt
```

```python
config_file = 'ssd_mobilenet_v3_large_coco_2020_01_14.pbtxt' # for configuring the file
frozen_model = 'frozen_inference_graph.pb' # for training the frozen model
```

```python
model = cv2.dnn_DetectionModel(frozen_model, config_file) # loaded the model
```

```python
classLabels = [] # empty list
file_name = 'Labels.txt' # Name of 80 classes
with open(file_name, 'rt') as fpt:   # to open a file
    classLabels = fpt.read().rstrip('\n').split('\n')
    # classLabels.append(fpt.read()) method adds a single item to the existing list
```

```python
print(classLabels)
```

```python
model.setInputSize(320, 320)
model.setInputScale(1.0/127.5) # 255/2 = 127.5
model.setInputMean((127.5, 127.5, 127.5)) # mobilenet => [-1, 1] # double first bracket (()) is used to act as 1 parameter
model.setInputSwapRB(True)
```

# Read an Image

```python
img = cv2.imread('Photos/man_bmw.png') 
```

```python
plt.imshow(img) # bgr (blue-green-red)
```

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))  # convert bgr to rgb
```

```python
ClassIndex, confidence, bbox = model.detect(img, confThreshold = 0.5)
```

```python
font_scale = 3 # formating the Text by scaling
font = cv2.FONT_HERSHEY_PLAIN # font family
for ClassInd, conf, boxes in zip(ClassIndex.flatten(), confidence.flatten(), bbox): #flatten() function we can flatten a matrix to one dimension in python
    # cv2.rectangle(frame, (x, y), (x+w, y+h), (255, 0, 0), 2)
    cv2.rectangle(img, boxes, (255, 0, 0), 2)
    
    # cv2.putText(img, text, (text_offset_x, text_offset_y), font, fontScale = font_scale, color = (0, 0, 0), thickness = 1)
    cv2.putText(img, classLabels[ClassInd-1], (boxes[0]+10, boxes[0]+40), font, fontScale = font_scale, color = (0, 255, 0), thickness = 3)
```

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))  # convert bgr to rgb
```

# Reading from Video

```python
video = cv2.VideoCapture("Videos/4K_Road_traffic_video_for_object_detection_and_tracking.mp4")
# Check if the video is opened correctly
if not video.isOpened():
    video = cv2.VideoCapture(0)
if not video.isOpened():
    raise IOError("Cannot open video")
    
font_scale = 3 # formating the Text by scaling
font = cv2.FONT_HERSHEY_PLAIN # font family
while True:
    ret, frame = video.read()
    ClassIndex, confidence, bbox = model.detect(frame, confThreshold = 0.55)
    
    if(len(ClassIndex) != 0):
        for ClassInd, conf, boxes in zip(ClassIndex.flatten(), confidence.flatten(), bbox): #flatten() function we can flatten a matrix to one dimension in python
            if(ClassInd <= 80):
                # cv2.rectangle(frame, (x, y), (x+w, y+h), (255, 0, 0), 2)
                cv2.rectangle(frame, boxes, (255, 0, 0), 2)

                # cv2.putText(frame, text, (text_offset_x, text_offset_y), font, fontScale = font_scale, color = (0, 0, 0), thickness = 1)
                cv2.putText(frame, classLabels[ClassInd-1], (boxes[0]+10, boxes[0]+40), font, fontScale = font_scale, color = (0, 255, 0), thickness = 3)
                
    cv2.imshow('Object Detection Tutorial', frame)
    
    if cv2.waitKey(2) & 0xFF == ord('q'):
        break
        
video.release()
cv2.destroyAllWindows()
```
